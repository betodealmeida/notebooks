We're mostly interested in extracting date/time, latitude, longitude and the pressure/temperature/salinity records.

```python
from datetime import datetime
import re

import numpy as np
from seawater import csiro

# regexp to extract date, time, lat and lon
# *DATE=27111997 TIME=1235 LAT=S23 50.60 LON=W041 20.35 DEPTH=1149   QC=1111
extract = re.compile("""
    DATE=(\d{8})\s
    TIME=(\d{4})\s
    LAT=((?:S|N)\d{2}\s\d{2}\.\d{2})\s
    LON=((?:W|E)\d{3}\s\d{2}\.\d{2})\s
""", re.VERBOSE)


def parse_latlon(value):
    if value[0] in 'SW':
        sign = -1
    else:
        sign = 1
    degrees, minutes = map(float, value[1:].split(' '))
    return sign*(degrees + minutes/60.)


casts = []

with open('97080060.ctd') as fp:
    for line in fp:
        if line.startswith('*DATE'):
            # found a cast
            date, time, lat, lon = extract.search(line).groups()
            cast = {
                    'datetime': datetime.strptime(date + time, '%d%m%Y%H%M'),
                    'latitude': parse_latlon(lat),
                    'longitude': parse_latlon(lon),
                    'data': [],
                    }
            casts.append(cast)
        elif line.startswith('  '):
            recs = map(float, line.strip().split(' '))
            cast['data'].append(recs[:3])
```
