```python
dataset = open_url("http://dap.marinexplore.org/dap/u/7b1af958cfd3/4abe3734-ff02-11e2-8d20-22000aeb03c4")
print dataset
```

```python
argo = np.rec.fromrecords(list(dataset.insitu), names=dataset.insitu.keys())
```

```python
bathy = dataset.g_bedrock_altitude.array[:][0,0]
lat0 = dataset.lat0[:]
lon0 = dataset.lon0[:]
```

```python
from mpl_toolkits.basemap import Basemap

fig = plt.figure(figsize=(8,8))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

m = Basemap(
    llcrnrlon=min(lon0),
    llcrnrlat=min(lat0),
    urcrnrlon=max(lon0),
    urcrnrlat=max(lat0),
    resolution='l',
    projection='merc')

X, Y = m(*np.meshgrid(lon0, lat0))
land = np.ma.masked_where(bathy<=0, bathy)
water = np.ma.masked_where(bathy>=0, bathy)
m.contourf(X, Y, land, levels=np.linspace(np.min(bathy), np.max(bathy), 41), cmap=cm.gist_earth, extend="both")
m.contourf(X, Y, water, levels=np.linspace(np.min(water), np.max(water), 41), cmap=cm.Blues_r, extend="both")
m.contour(X, Y, bathy, levels=[0,], colors="k", linewidth=0.5)
cs = m.contour(X, Y, bathy, levels=[-50, -1000], colors="k", linewidth=0.1)
plt.clabel(cs, inline=1, fontsize=10, fmt="%d")
for id_ in np.unique(argo["stid"]):
    platform = argo[argo["stid"] == id_]
    x, y = m(platform["lat"], platform["lon"])
    m.plot(x, y, color="r", alpha=0.5)
    m.scatter(x[-1:], y[-1:], 50, color="r", edgecolors="#660000")
plt.show()
```
