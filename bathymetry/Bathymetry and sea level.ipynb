```python
from pupynere import *

f = netcdf_file("/usr/local/ferret/datasets/data/etopo5.cdf")
bathy = f.variables["ROSE"][:]
LAT, LON = f.variables["ROSE"].dimensions
lat = f.variables[LAT][:]
lon = f.variables[LON][:]
```

```python
from mpl_toolkits.basemap import Basemap

fig = plt.figure(figsize=(12,12))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

m = Basemap(
    llcrnrlon=min(lon),
    llcrnrlat=-80,
    urcrnrlon=max(lon),
    urcrnrlat=85,
    resolution='l',
    projection='merc')
m = Basemap(projection='robin',lon_0=lon.mean(),resolution='h')

X, Y = m(*np.meshgrid(lon, lat))
land = np.ma.masked_where(bathy<=0, bathy)
water = np.ma.masked_where(bathy>=0, bathy)
m.contourf(X, Y, land, levels=np.linspace(np.min(bathy), np.max(bathy), 41), cmap=cm.gist_earth, extend="both")
m.contourf(X, Y, water, levels=np.linspace(np.min(water), np.max(water), 41), cmap=cm.Blues_r, extend="both")
#.drawcoastlines(linewidth=0.5)
m.contour(X, Y, bathy, levels=[0], colors="k", linewidth=0.5)
ax.set_title('Current sea level')
plt.show()
```

```python
from mpl_toolkits.basemap import Basemap

fig = plt.figure(figsize=(12,12))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

m = Basemap(
    llcrnrlon=250,
    llcrnrlat=10,
    urcrnrlon=300,
    urcrnrlat=40,
    resolution='l',
    projection='merc')

n = len(lon)
lon2 = np.concatenate((lon[n/2:], lon[:n/2]+360), axis=0)
bathy2 = np.concatenate((bathy[:,n/2:], bathy[:,:n/2]), axis=1)
X, Y = m(*np.meshgrid(lon2, lat))

flooded = np.ma.masked_where((bathy2>0.69) | (bathy2 <0), bathy2)
m.contourf(X, Y, flooded, cmap=cm.Reds, extend="both")
m.contour(X, Y, bathy2, levels=[0], colors="k", linewidth=0.5)
ax.set_title('Flooded areas in 2100')
plt.show()
```
