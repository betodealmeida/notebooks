A first look at HFR/CODAR data
==============================

I created a [prepared dataset](http://marinexplore.org/explore/#downloads/bc903c7e-fb95-11e2-8118-22000aeb03c4), containing surface current speeds at the Monterey Bay from 2000-08-02 to 2000-09-26. Let's access it using [Pydap](http://pydap.org/) (I'm using the development version of Pydap from the [repository](http://github.com/robertodealmeida/pydap)):

```python
from pydap.client import open_url

dataset = open_url("http://dap.marinexplore.org/dap/u/7b1af958cfd3/bc903c7e-fb95-11e2-8118-22000aeb03c4")
print dataset
```

```python
U = dataset.g_sea_water_current_vector_u.g_sea_water_current_vector_u[:][:,0]
U = np.ma.masked_where(isnan(U), U)

V = dataset.g_sea_water_current_vector_v.g_sea_water_current_vector_v[:][:,0]
V = np.ma.masked_where(isnan(V), V)

x = dataset.lon0[:]
y = dataset.lat0[:]

# calculate speed
Z = np.sqrt(U**2 + V**2)
```

```python
from mpl_toolkits.basemap import Basemap

fig = figure(figsize=(12,6))
ax = fig.gca()
m = Basemap(ax=ax,
    llcrnrlon=min(x),
    llcrnrlat=min(y),
    urcrnrlon=max(x)+0.3,
    urcrnrlat=max(y),
    projection="merc", resolution="h")
m.drawmapboundary(fill_color="0.3")
m.drawcoastlines(linewidth=1.25)
m.fillcontinents(color="0")

X, Y = m(*np.meshgrid(x, y))
colors = m.contourf(X, Y, np.mean(Z, axis=0), alpha=0.5)
m.streamplot(X, Y, np.mean(U, axis=0), np.mean(V, axis=0), color="white")
cb = colorbar(colors)
cb.set_label("m/s")

fig.savefig("mbari-speed.png")
```

Let's now calculate where we can find fish.

> **Convergence vs Divergence**
>
> (...) Common usage for the surface of the ocean is that convergence is the accumulation of water particles, and water sinks or is down-welled. Weed lines, that are great for fishing, and oceanic surface drifters are usually found in convergence zones. Divergence of surface waters is indicated by surface drifters leaving an area. Upwelled water, that is usually colder and nutrient-rich, from below replaces the diverging surface flow. [[ref](http://oceancurrents.rsmas.miami.edu/glossary.html)]

```python
xd = (x[:-1] + x[1:])/2.
yd = (y[:-1] + y[1:])/2.

Um = (U[:,:-1,:] + U[:,1:,:])/2.
Xm = (X[:-1,:] + X[1:,:])/2.
Ud = (Um[:,:,1:] - Um[:,:,:-1])/(Xm[:,1:] - Xm[:,:-1])

Vm = (V[:,:,:-1] + V[:,:,1:])/2.
Ym = (Y[:,:-1] + Y[:,1:])/2.
Vd = (Vm[:,1:,:] - Vm[:,:-1,:])/(Ym[1:,:] - Ym[:-1,:])

div = np.mean(Ud + Vd, axis=0)
```

```python
from mpl_toolkits.basemap import Basemap

fig = figure(figsize=(12,6))
ax = fig.gca()
m = Basemap(ax=ax,
    llcrnrlon=min(x),
    llcrnrlat=min(y),
    urcrnrlon=max(x)+0.3,
    urcrnrlat=max(y),
    projection="merc", resolution="l")
m.drawmapboundary(fill_color="0.3")
m.drawcoastlines(linewidth=1.25)
m.fillcontinents(color="0")

Xd, Yd = m(*np.meshgrid(xd, yd))
levels = np.linspace(-4e-5, 4e-5, 11) * 1e5
m.contour(Xd, Yd, div, np.linspace(-1, 1, 3), colors="1")
m.streamplot(X, Y, np.mean(U, axis=0), np.mean(V, axis=0), color="black")
colors = m.contourf(Xd, Yd, -div*1e5, levels, extend="both", alpha=1.0, cmap=cm.RdBu_r)
cb = colorbar(colors)
cb.set_label(u"divergence (blue) × convergence (red)")

fig.savefig("mbari-convergence.png")
```
