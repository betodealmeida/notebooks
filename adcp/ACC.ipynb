Let's start looking at [ADCP data measuring the Antarctice Circumpolar Current](http://rel-0-10.marinexplore.org/s/NdpmiJM). Here's a [prepared dataset](http://rel-0-10.marinexplore.org/explore/#downloads/57f33f4a-0609-11e3-b4a9-22000a9e3d51) with current data from the JASADCP project:

```python
adcp = np.genfromtxt("platform3/data.csv", delimiter=",", dtype=None, names=True, usemask=True)

from netCDF4 import Dataset
elevation = Dataset("NOAA.ETOPO1.BEDROCK.nc")
```

```python
print adcp.dtype.names
print elevation.variables
```

```python
bathy = elevation.variables["bedrock.altitude"][0,0,:]
bathy = np.ma.masked_where(isnan(bathy), bathy)
lon = elevation.variables["lon"][:]
lat = elevation.variables["lat"][:]
```

```python
from mpl_toolkits.basemap import Basemap

fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([0, 0, 1, 1])

m = Basemap(
    llcrnrlon=min(lon),
    llcrnrlat=min(lat),
    urcrnrlon=max(lon),
    urcrnrlat=max(lat),
    resolution='l',
    projection='merc')

# plot basemap
X, Y = m(*np.meshgrid(lon, lat))
land = np.ma.masked_where(bathy<=0, bathy)
water = np.ma.masked_where(bathy>=0, bathy)
m.contourf(X, Y, land, levels=np.linspace(np.min(bathy), np.max(bathy), 41), cmap=cm.gist_earth, extend="both")
m.contourf(X, Y, water, levels=np.linspace(np.min(water), np.max(water), 41), cmap=cm.Blues_r, extend="both")

# plot ship tracks

for id_ in np.unique(adcp["station_id"]):
    cruise = adcp[adcp["station_id"] == id_]
    y = cruise["longitude_degrees_east"]
    x = cruise["latitude_degrees_north"]
    X, Y = m(x, y)
    m.plot(X, Y, 'r-')

ax.set_title('Ship track')
ax.set_axis_off()
plt.show()
```

Now let's put the data into a time x depth matrix:

```python
def distance(lat1, lon1, lat2, lon2):
    R = 6.371e6
    lat1 *= pi/180.
    lon1 *= pi/180.
    lat2 *= pi/180.
    lon2 *= pi/180.
    return R*np.arccos(
        np.sin(lat1)*np.sin(lat2) +
        np.cos(lat1)*np.cos(lat2)*np.cos(lon2-lon1))
```

```python
z = np.unique(adcp["depth_meters"])
t = np.unique(adcp["date_time"])

U = np.ma.masked_all((len(t), len(z)), "d")
V = np.ma.masked_all((len(t), len(z)), "d")
x = np.ma.masked_all((len(t),), "d")
y = np.ma.masked_all((len(t),), "d")
for rec in cruise:
    k = (z == rec["depth_meters"]).nonzero()[0][0]
    l = (t == rec["date_time"]).nonzero()[0][0]
    U[l,k] = rec["sea_watercurrentspeedu_ms"]
    V[l,k] = rec["sea_watercurrentspeedv_ms"]
    x[l] = rec["latitude_degrees_north"]
    y[l] = rec["longitude_degrees_east"]
    
from datetime import datetime
import coards
t = [datetime.strptime(v, "%Y-%m-%dT%H:%M:%SZ") for v in t]
t = [coards.format(d, "days since 2010-12-31") for d in t]
t = np.array(t)
```

```python
distances = distance(y[:-1].copy(), x[:-1].copy(), y[1:].copy(), x[1:].copy())
dz = z[1:] - z[:-1]
Ua = (U[:-1,:].filled(0) + U[1:,:].filled(0))/2.
Ua = (Ua[:,:-1] + Ua[:,1:])/2.
area = distances[:,None] * dz[None,:]
transport = area * Ua

# get only the southward leg
print sum(transport[:29])/1e6
```

What if extend the current to the bottom?

```python
depth = np.ma.masked_all((len(t),), "d")
for l, (x0, y0) in enumerate(zip(x, y)):
    i = abs(lon-x0).argmin()
    j = abs(lat-y0).argmin()
    depth[l] = bathy[j,i]
depth = (depth[:-1] + depth[1:])/2.
    
Ua = (U[:-1,:].filled(0) + U[1:,:].filled(0))/2.
transport = distances*-depth*mean(Ua, axis=1)
print sum(transport[:29])/1e6
```

```python
def bearing(lat1, lon1, lat2, lon2):
    # http://www.movable-type.co.uk/scripts/latlong.html
    lat1 *= pi/180.
    lon1 *= pi/180.
    lat2 *= pi/180.
    lon2 *= pi/180.
    y = np.sin(lon2-lon1)*np.cos(lat2)
    x = np.cos(lat1)*np.sin(lat2) - np.sin(lat1)*np.cos(lat2)*np.cos(lon2-lon1)
    return np.arctan2(y, x)
```
