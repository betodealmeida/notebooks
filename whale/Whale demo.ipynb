```python
from scipy.io import wavfile
from scipy import signal
import numpy as np
```

```python
whale = []
nowhale = []

for sample in range(5):
    rate, x = wavfile.read('small_data_sample/whale%d.wav' % (sample+1))
    whale.append(x)
    
    rate, x = wavfile.read('small_data_sample/nowhale%d.wav' % (sample+1))
    nowhale.append(x)
```

```python
fig = plt.figure(figsize=(12, 24))
for sample in range(5):
    ax = plt.subplot(5, 2, sample*2+1)
    ax.specgram(whale[sample], Fs=rate)
    if sample == 0:
        ax.set_title('whale')
        
    ax = plt.subplot(5, 2, sample*2+2)
    ax.specgram(nowhale[sample], Fs=rate)
    if sample == 0:
        ax.set_title('nowhale')
```

```python
fig = plt.figure(figsize=(10, 6.18))

ax = plt.subplot(3, 1, 1)
ax.plot(np.arange(len(nowhale[4]), dtype='f')/rate, nowhale[4])
ax.set_title('Whale up-call')
ax.axis((0.064, 1.92, -1500, 1500))
ax.set_ylabel('Intensity')

ax = plt.subplot(3, 1, 2)
ax.specgram(nowhale[4], Fs=rate)
ax.axis((0.064, 1.92, 100, 500))
ax.set_ylabel('frequency (Hz)')

ax = plt.subplot(3, 1, 3)
P, y, x = plt.mlab.specgram(nowhale[4], Fs=rate)
ax.contourf(x, y, P, 10)
ax.axis((0.064, 1.92, 100, 500))
ax.set_xlabel('time (s)')
ax.set_ylabel('frequency (Hz)')
```

```python
def bandpass(timeseries):
    fso2 = rate/2.
    N, wn = signal.buttord(ws=[100/fso2, 250/fso2], wp=[50/fso2, 300/fso2], gpass=-1.0, gstop=30.0)

    b, a = signal.butter(N, wn, btype='band')
    return signal.lfilter(b, a, timeseries)
```

```python
fig = plt.figure(figsize=(12, 24))
for sample in range(5):
    ax = plt.subplot(5, 2, sample*2+1)
    ax.specgram(bandpass(whale[sample]), Fs=rate)
    ax.axis((0, 2, 0, 500))
    if sample == 0:
        ax.set_title('whale')
        
    ax = plt.subplot(5, 2, sample*2+2)
    ax.specgram(bandpass(nowhale[sample]), Fs=rate)
    ax.axis((0, 2, 0, 500))
    if sample == 0:
        ax.set_title('nowhale')
```

```python
fig = plt.figure(figsize=(12, 12))
for sample in range(5):
    ax = plt.subplot(5, 2, sample*2+1)
    P, y, x = plt.mlab.specgram(bandpass(whale[sample]))
    ax.pcolor(x, y, np.diff(np.diff(P, axis=0), axis=1))
    print 'whale', P.max()
    if sample == 0:
        ax.set_title('whale')
        
    ax = plt.subplot(5, 2, sample*2+2)
    P, y, x = plt.mlab.specgram(bandpass(nowhale[sample]))
    ax.pcolor(x, y, np.diff(np.diff(P, axis=0), axis=1))
    print 'nowhale', P.max()
    if sample == 0:
        ax.set_title('nowhale')
```
